```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

SEC_FACTOR = 1e+6
```

```python
w_2_c_1_p_100 = pd.read_csv("benchmark_2_workers_1_CPU_100.csv")
w_4_c_1_p_50 = pd.read_csv("benchmark_4_workers_1_CPU_50.csv")
w_4_c_1_p_100 = pd.read_csv("benchmark_4_workers_1_CPU_100.csv")
w_8_c_1_p_50 = pd.read_csv("benchmark_8_workers_1_CPU_50.csv")
w_8_c_1_p_100 = pd.read_csv("benchmark_8_workers_1_CPU_100.csv")
w_8_c_2_p_100 = pd.read_csv("benchmark_8_workers_2_CPU_100.csv")
```

```python
w_2_c_1_p_100.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
w_4_c_1_p_50.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
w_4_c_1_p_100.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
w_8_c_1_p_50.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
w_8_c_1_p_100.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
w_8_c_2_p_100.sort_values("ID", axis=0, ascending=True, inplace=True, na_position="last")
```

```python
xpoints = np.array(w_2_c_1_p_100.ID)
ypoints = np.array(w_2_c_1_p_100.Time / SEC_FACTOR)

plt.plot(xpoints, ypoints)

plt.title(f"Workers - 2, CPU - 1, CPU Percentage - 100, Average - {ypoints.sum() / 100}")
plt.xlabel("Jobs")
plt.ylabel("Milliseconds")

plt.show()
```
